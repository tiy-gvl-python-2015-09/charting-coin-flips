```python
def new_coin_flip():
    # heads = 1 tails = 0
    return choice(["1", "0"])

def one_hundred_flips():
    hundred = []
    for _ in range(100):
        hundred.append(new_coin_flip())
    return hundred

def one_thousand_flips():
    thousand = []
    for _ in range(1000):
        thousand.append(new_coin_flip())
    return thousand
```

```python
def hundred_thousand():
    large_number = []
    for _ in range(100000):
        large_number.append(one_hundred_flips())
    return large_number
```

```python
def hundred_thousand2():
    large_number = []
    for _ in range(100000):
        large_number.append(one_thousand_flips())
    return large_number
```

```python
trials = hundred_thousand()
```

```python
heads_ratio = []
for trial in trials:
    heads_ratio.append(trial.count("1")/len(trial))
```

```python
plt.hist(heads_ratio, bins = 25)
```

In this histogram we can see that the ratio of heads to total amount of coin flips in 100 coins flipped over 100,000 trials is almost a gaussian distribution, the hight of which is just around 50%. While there are some outliers, the main thing to infer from this is that chances are you will have a 50/50 chance of getting heads in a coin toss.
